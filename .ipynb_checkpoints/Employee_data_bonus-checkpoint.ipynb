```python
# Dependencies
import pandas as pd
import psycopg2
import matplotlib.pyplot as plt
import numpy as np

from config import api_key

```

```python
from sqlalchemy import create_engine
```

```python
DB_URI = 'postgres+psycopg2://postgres:'+api_key+'@localhost:5432/employees_db'
engine = create_engine(DB_URI)
conn = engine.connect()
```

```python
# Selecting the Salaries table schema and creating a DF
salary = pd.read_sql("SELECT * FROM Salaries", conn)

salary.head(10)
```

```python
#remove special characters in salary column
def clean_salary(salary):
    return float(salary.replace('$','').replace(',',''))
```

```python
clean_salary('$60,929.00') 
```

```python
x = salary["salary"].map(clean_salary)
x[:10]
```

```python
#Plotting with pyplot
plt.figure(figsize=(8, 6), dpi=70) # Figure size
plt.hist(x, facecolor='b', alpha=0.75)

# Aesthetics for the chart
plt.title("Most Common Employee Salary Ranges ", fontsize=14)
plt.xlabel("Salary ranges($)", fontsize=12)
plt.ylabel("Frequency - Employees", fontsize=12)
plt.grid(5000)
plt.xlim(35000, 125000)
plt.show()
```

```python
title = pd.read_sql("SELECT * FROM Titles", conn)

title
```

```python
# Merge of the DFs based on emp_no
salary_title = pd.merge(title, salary, on = ["emp_no"], how="inner")

#Displaying the merged DF
salary_title
```

```python
#To remove $ and commas "," - acces the salary column in the salary_title DF, map() pass a function on every cell 
#in the DF and 
sal = salary_title["salary"].map(clean_sal)
sal[:5]
```

```python
sal_title = pd.DataFrame({"employee tit":salary_tit["title"], "salary":sal})
sal_title
```

```python
group_sal_title = sal_title.groupby(['employee tit'])
group_sal_title.count()
```

```python
av_sal = group_sal_title["salary"].mean()
av_sal
```

```python
#Creating a DF with the results

sal_title_DF = pd.DataFrame({"Average Salary":av_sal})
sal_title_DF
```

```python
#Variables
x = sal_title_DF.index
y = sal_title_DF["Average Salary"]

#Plotting with pyplot
plt.figure(figsize=(8, 6), dpi=70) # Figure size
plt.bar(x, y, color='r', align="center", width=0.6)

# Aesthetics for the chart
plt.xticks(x,rotation=90)
plt.title("Average Employee Salary by Title", fontsize=14)
plt.xlabel("Average Salary($)", fontsize=12)
plt.ylabel("Employee Title", fontsize=12)
plt.show()
```
